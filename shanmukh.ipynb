```python
import matplotlib.pyplot as plt

categories=['trabsport','income tax','housing']
values1=[15,24,30]
values2=[20,18,25]

plt.bar(categories,values1,label='first half',color='green')
plt.bar(categories,values2,bottom= values1,label='second half',color='orange')
plt.xlabel("categories")
plt.ylabel('values')
plt.title('stacked Bar Graph')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories=['lion','Horse','kangarooo']
values1=[15,24,30]
values2=[20,18,25]

bar_width=0.35
bar_positions1=np.arange(len(categories))
bar_positions2=bar_positions1+bar_width

plt.bar(bar_positions1,values1,width=bar_width,label='2021',color='skyblue')
plt.bar(bar_positions2,values2,width=bar_width,label='2022',color='green')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('ground bar graph')
plt.legend()
plt.show()
```

```python
import pandas as pd
data1={'Name ':['jai','princi','gaurav','Anju'],
        'Address':['Nagpur','Kanpur','Allahabad','Kannuaj'],
        'Age':[27,24,22,32],
        'Qualification':['Msc','Mca','MA','PHD']}
data2={'Name ':['Abhi','Ayushi','dhiraj','hekash'],
        'Address':['Nagpur','Kanpur','Allahabad','Kannuaj'],
        'Age':[17,14,12,52],
        'Qualification':['B.Tech','B.com','B.hons','B.A']}

df=pd.DataFrame(data1,index=[0,1,2,3])
df1=pd.DataFrame(data2,index=[2,3,6,7])
print(df,"\n\n",df1)
res=pd.concat([df,df1],ignore_index=True)
res
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import read_csv

path="/content/drive/MyDrive/energy1.csv"
electricity=pd.read_csv(path)

print(electricity)
print(electricity.shape)
print(electricity.head(5))
print(electricity.tail(4))
print(electricity.info())
print(electricity.describe())

print("column Data")
print(electricity.columns)
print(electricity[electricity['State'].str.startswith('M')])
print(electricity[electricity['State'].str.endswith('a')])
rcb=electricity[electricity['State']=='Andhra']
print(len(rcb))

print(electricity.sort_values(by='State'))

print(np.max(electricity['Consumption']))
print(np.min(electricity['Consumption']))
print(np.mean(electricity['Consumption']))


state1_data=electricity[electricity.State=='Andhra']
print("Andhra Data")
print(state1_data)
state2_data=electricity[electricity['State']=='Karnataka']
state4_data=electricity[electricity['State']=='Telengana']
state5_data=electricity[electricity['State']=='Tripura']

fig,axes=plt.subplots(2)
axes[0].plot(state1_data['Date'],state1_data['Consumption'],marker='*',color='red')
axes[1].plot(state2_data['Date'],state1_data['Consumption'],marker='^',color='green')
#bar chart
fig1,axes1=plt.subplots(2)
axes1[0].bar(state4_data['Date'],state1_data['Consumption'],color='yellow')
axes1[1].bar(state5_data['Date'],state1_data['Consumption'],color='pink')


#pie chart
fig2,axes2=plt.subplots(2)
state1='UP'
state2='Kerala'
state3='Mizoram'
state4='Delhi'

labels=[state1,state2]
state1_total=electricity[electricity['State']==state1]['Consumption'].sum()
state2_total=electricity[electricity['State']==state2]['Consumption'].sum()
# 'energy' should be a list of numerical values
energy=[state1_total,state2_total]
axes2[0].pie(energy,labels=labels,autopct='%1.1f%%')

labels=[state3,state4]
state3_total=electricity[electricity['State']==state3]['Consumption'].sum()
state4_total=electricity[electricity['State']==state4]['Consumption'].sum()
# 'energy' should be a list of numerical values
energy=[state3_total,state4_total]
axes2[1].pie(energy,labels=labels,autopct='%1.1f%%')





```
